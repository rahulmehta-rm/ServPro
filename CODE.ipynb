```python
import random

# Randomize rows while ensuring no two consecutive rows have the same 'Location'

# Step 1: Shuffle the DataFrame rows
final_data_shuffled = final_data.sample(frac=1, random_state=42).reset_index(drop=True)

# Step 2: Check for consecutive duplicates in 'Location' and reshuffle until no consecutive duplicates exist
# (This is a simplistic approach that might need to be adjusted based on your data)
def has_consecutive_duplicates(df, column):
    return any(df[column].iloc[i] == df[column].iloc[i + 1] for i in range(len(df) - 1))

# Keep shuffling until no consecutive duplicates in 'Location' are found
attempt = 0
while has_consecutive_duplicates(final_data_shuffled, 'Location') and attempt < 10:  # Limit to 10 attempts to avoid infinite loop
    final_data_shuffled = final_data_shuffled.sample(frac=1).reset_index(drop=True)
    attempt += 1

# Display the randomized DataFrame to verify
print(final_data_shuffled[['Location']].head(10))
```

```python
def recommend_top_5_services(location, service_type):
    """
    Function to recommend the top 5 service providers in a given city for a specified service type.
    
    Parameters:
    location (str): The city or area where services are requested.
    service_type (str): The type of service (e.g., 'Plumbing', 'Water Heater Installation/Repair').
    
    Returns:
    DataFrame: Top 5 service providers sorted by rating.
    """
    # Filter dataset by location
    location_filtered = final_data[final_data['Location'].str.contains(location, case=False, na=False)]
    
    # Further filter by service type
    if service_type in location_filtered.columns:
        service_filtered = location_filtered[location_filtered[service_type] == 1]
    else:
        print(f"Service type '{service_type}' not found in dataset.")
        return pd.DataFrame()
    
    # Sort by rating in descending order and pick the top 5
    top_5 = service_filtered.sort_values(by='Rating', ascending=False).head(5)
    
    # Return the top 5 service providers with relevant columns
    return top_5[['Name', 'Rating', 'Address','Location', 'Services Provided', 'Phone Number', 'Website']]

# Example: Recommend top 5 plumbers in 'Scarborough'
top_5_plumbers_scarborough = recommend_top_5_services(location='Hamilton', service_type='Plumbing')
top_5_plumbers_scarborough
```

```python
# Normalize the 'Location' column
def clean_location(location):
    # Remove extra information after a comma (e.g., postal codes, province)
    location = re.split(',', location)[0].strip().lower()
    
    # Correct common misspellings and inconsistent spellings
    corrections = {
        'bramton': 'brampton',
        'kitchner': 'kitchener',
        'vaughan': 'vaughan'
    }
    if location in corrections:
        location = corrections[location]
    
    return location.capitalize()

# Apply the cleaning function to the 'Location' column
final_data['Location'] = final_data['Location'].apply(clean_location)

# Verify the unique values after cleaning
final_data['Location'].unique()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier

# Encode the 'Location' and 'Services Provided' columns for use in the model
le_location = LabelEncoder()
final_data['Location_Encoded'] = le_location.fit_transform(final_data['Location'])

le_services = LabelEncoder()
final_data['Service_Encoded'] = le_services.fit_transform(final_data['Services Provided'])

# Define features (X) and target (y)
X = final_data[['Location_Encoded', 'Rating', 'Number of Reviews']]
y = (final_data['Rating'] >= 4.5).astype(int)  # Target: 1 if the service is highly rated (>= 4.5), else 0

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=18)

# Display the size of training and testing sets
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize Logistic Regression
lr_model = LogisticRegression(random_state=18, max_iter=750)

# Train the model
lr_model.fit(X_train, y_train)

# Make predictions
y_pred_lr = lr_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_lr)
precision = precision_score(y_test, y_pred_lr, average='weighted')
recall = recall_score(y_test, y_pred_lr, average='weighted')
f1 = f1_score(y_test, y_pred_lr, average='weighted')

print("Logistic Regression Model Performance:")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

# Implement k-Nearest Neighbors (kNN) Model for Recommendation

```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load the dataset
file_path = 'Final_Merged_Data.csv'
final_data = pd.read_csv(file_path)

# Step 1: Data Preprocessing

# Normalize the 'Location' column
def clean_location(location):
    # Remove extra information after a comma (e.g., postal codes, province)
    location = re.split(',', location)[0].strip().lower()
    
    # Correct common misspellings and inconsistent spellings
    corrections = {
        'bramton': 'brampton',
        'kitchner': 'kitchener',
        'vaughan': 'vaughan'
    }
    if location in corrections:
        location = corrections[location]
    
    return location.capitalize()

# Apply the cleaning function to the 'Location' column
final_data['Location'] = final_data['Location'].apply(clean_location)

# Fill missing values
final_data['Opening Hours'] = final_data['Opening Hours'].fillna('Not Provided')
final_data['Response Time'] = final_data['Response Time'].fillna('Not Provided')
final_data['Response Rate'] = final_data['Response Rate'].fillna('Not Provided')
final_data['Phone Number'] = final_data['Phone Number'].fillna('Unavailable')
final_data.dropna(subset=['Rating'], inplace=True)

# One-Hot Encode 'Services Provided'
services_one_hot = final_data['Services Provided'].str.get_dummies(sep=', ')
final_data = pd.concat([final_data, services_one_hot], axis=1)

# Label Encode the 'Location' and 'Services Provided' columns
le_location = LabelEncoder()
final_data['Location_Encoded'] = le_location.fit_transform(final_data['Location'])

le_services = LabelEncoder()
final_data['Service_Encoded'] = le_services.fit_transform(final_data['Services Provided'])

# Define features (X) and target (y)
X = final_data[['Location_Encoded', 'Rating', 'Number of Reviews']]
y = (final_data['Rating'] >= 4.5).astype(int)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=18)

# Standardize the features for kNN model
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Display the size of training and testing sets
X_train.shape, X_test.shape
```

```python
# Initialize the kNN Classifier with 5 neighbors
knn_model = KNeighborsClassifier(n_neighbors=5)

# Train the kNN model on the scaled training data
knn_model.fit(X_train_scaled, y_train)

# Make predictions on the test data
y_pred_knn = knn_model.predict(X_test_scaled)

# Evaluate the kNN model using accuracy, precision, recall, and F1 score
accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn)
recall_knn = recall_score(y_test, y_pred_knn)
f1_knn = f1_score(y_test, y_pred_knn)

print(f"kNN Model Performance:")
print(f"Accuracy: {accuracy_knn:.2f}")
print(f"Precision: {precision_knn:.2f}")
print(f"Recall: {recall_knn:.2f}")
print(f"F1 Score: {f1_knn:.2f}")
```
