```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("Final_Merged_Data.csv")
```

```python
# Find rows with duplicate 'Name' values
duplicate_names = df[df.duplicated(subset='Name', keep=False)]

# Group by 'Name' and count the occurrences
name_counts = duplicate_names['Name'].value_counts()

print("Count of rows with the same name:")
name_counts
```

```python
df_cleaned = df.drop_duplicates(subset='Name')
print("Duplicates based on 'Name' removed. Cleaned DataFrame:")
df_cleaned.head()
```

```python
unique_count = df['Name'].nunique()
print(f"Number of unique values in 'Name' column: {unique_count}")
```
